```python
# The PyReft library is integrated to establish the ReFT baseline. The evaluation of ReFT is prioritized for tasks like GSM8K and RAFT to test the hypothesis that internal representation interventions capture reasoning chains more efficiently than weight updates.

# Version pinning for environment consistency.
!pip install transformers==4.44.2 trl==0.9.6 accelerate==0.34.2 peft==0.12.0 bitsandbytes datasets huggingface_hub fpdf pyreft
!pip install --force-reinstall charset-normalizer

import time, re, torch, textwrap, pyreft, shutil, os, gc
import matplotlib.pyplot as plt
from datasets import load_dataset, Dataset
from transformers import AutoModelForCausalLM, AutoTokenizer, TrainingArguments, BitsAndBytesConfig
from peft import LoraConfig, get_peft_model, TaskType, prepare_model_for_kbit_training
from trl import SFTTrainer
from fpdf import FPDF
from google.colab import drive
from huggingface_hub import login

# --- PYREFT COMPATIBILITY PATCH ---
from pyreft import ReftTrainerForCausalLM
_orig_compute_loss = ReftTrainerForCausalLM.compute_loss
def _patched_compute_loss(self, model, inputs, return_outputs=False, **kwargs):
    kwargs.pop("num_items_in_batch", None)
    return _orig_compute_loss(self, model, inputs, return_outputs=return_outputs)
ReftTrainerForCausalLM.compute_loss = _patched_compute_loss

# --- INITIALIZATION ---
drive.mount('/content/drive', force_remount=True)
HF_TOKEN = "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"
login(token=HF_TOKEN)

COMPUTE_DTYPE = torch.bfloat16 if torch.cuda.is_bf16_supported() else torch.float16
USE_BF16, USE_FP16 = (True, False) if torch.cuda.is_bf16_supported() else (False, True)

class DataProcessor:
    @staticmethod
    def load_all():
        datasets = {}
        load_configs = {
            "gsm8k": ("openai/gsm8k", "main", "train"),
            "superglue": ("super_glue", "boolq", "validation"),
            "fin_sentiment": ("financial_phrasebank", "sentences_allagree", "train"),
            "raft": ("ought/raft", "ade_corpus_v2", "train"),
            "ifbench": ("Muennighoff/IFEval", None, "train")
        }
        for name, cfg in load_configs.items():
            try:
                datasets[name] = load_dataset(cfg[0], cfg[1], split=cfg[2]) if cfg[1] else load_dataset(cfg[0], split=cfg[2])
            except: pass
        return datasets

    @staticmethod
    def get_fields(name, sample):
        if name == "gsm8k": return sample["question"], sample["answer"]
        if name == "superglue": return f"{sample['passage']}\nQ: {sample['question']}", str(sample["label"])
        if name == "fin_sentiment": return sample["sentence"], str(sample["label"])
        if name == "raft": return sample["Sentence"], str(sample["Label"])
        return sample.get("instruction", ""), str(sample.get("response", ""))

    @staticmethod
    def calculate_accuracy(pred, truth):
        return 1.0 if truth.strip().lower() in pred.strip().lower() else 0.0

class BaselineExecutor:
    def __init__(self, model, tokenizer):
        self.model, self.tokenizer = model, tokenizer

    def run(self, query, context):
        self.model.train()
        self.model.enable_input_require_grads()

        ds = Dataset.from_dict({"text": [f"{ex}{self.tokenizer.eos_token}" for ex in context]})
        config = LoraConfig(r=8, lora_alpha=16, target_modules=["q_proj", "v_proj"], task_type=TaskType.CAUSAL_LM)
        pm = get_peft_model(self.model, config)

        args = TrainingArguments(
            output_dir="./l",
            per_device_train_batch_size=1,
            gradient_accumulation_steps=4,
            num_train_epochs=1,
            fp16=USE_FP16,
            bf16=USE_BF16,
            report_to="none",
            gradient_checkpointing=True
        )
        trainer = SFTTrainer(model=pm, train_dataset=ds, dataset_text_field="text", tokenizer=self.tokenizer, args=args)
        trainer.train()

        pm.eval()
        inputs = self.tokenizer(query, return_tensors="pt").to("cuda")
        with torch.no_grad():
            # Ensure proper unpacking of tokenizer inputs
            out = pm.generate(**inputs, max_new_tokens=20)
        res = self.tokenizer.decode(out[0][inputs['input_ids'].shape[1]:], skip_special_tokens=True)

        del trainer, pm; torch.cuda.empty_cache(); gc.collect()
        return res

class ReFTExecutor:
    def __init__(self, model, tokenizer):
        self.model, self.tokenizer = model, tokenizer

    def run(self, query, context):
        self.model.train()
        self.model.enable_input_require_grads()
        self.model.gradient_checkpointing_disable()
        self.model.config.use_cache = False

        reft_config = pyreft.ReftConfig(representations=[{
            "layer": 15, "component": "block_output", "low_rank_dimension": 4,
            "intervention": pyreft.LoreftIntervention(embed_dim=self.model.config.hidden_size, low_rank_dimension=4, dtype=COMPUTE_DTYPE)
        }])
        rm = pyreft.get_reft_model(self.model, reft_config)
        rm.set_device("cuda")

        prompts = [ex.rsplit("Output:", 1)[0] + "Output:" for ex in context]
        targets = [ex.rsplit("Output:", 1)[1].strip() if "Output:" in ex else "" for ex in context]

        dm = pyreft.make_last_position_supervised_data_module(self.tokenizer, self.model, prompts, targets)

        args = TrainingArguments(
            output_dir="./r",
            per_device_train_batch_size=1,
            gradient_accumulation_steps=4,
            num_train_epochs=1,
            fp16=USE_FP16,
            bf16=USE_BF16,
            report_to="none",
            remove_unused_columns=False,
            gradient_checkpointing=False
        )

        trainer = pyreft.ReftTrainerForCausalLM(model=rm, tokenizer=self.tokenizer, args=args, **dm)
        trainer.train()

        rm.eval()
        inputs = self.tokenizer(query, return_tensors="pt").to("cuda")
        position = inputs["input_ids"].shape[1] - 1
        unit_locations = {"sources->base": (None, [[[position]]])}

        with torch.no_grad():
            # Inputs are unpacked and autocast is used to maintain dtype consistency
            with torch.autocast("cuda", dtype=COMPUTE_DTYPE):
                _, out = rm.generate(
                    inputs, # pyreft expects the BatchEncoding object directly or input_ids
                    unit_locations=unit_locations,
                    intervene_on_prompt=True,
                    max_new_tokens=20,
                    do_sample=False,
                    eos_token_id=self.tokenizer.eos_token_id,
                    early_stopping=True
                )
        res = self.tokenizer.decode(out[0][inputs['input_ids'].shape[1]:], skip_special_tokens=True)

        self.model.config.use_cache = True
        del rm, trainer; torch.cuda.empty_cache(); gc.collect()
        return res

def generate_performance_chart(results_data, output_path):
    datasets = list(results_data.keys())
    n_values = [16, 32, 64, 128, 256]

    fig, axes = plt.subplots(len(datasets), 1, figsize=(10, 5 * len(datasets)))
    if len(datasets) == 1: axes = [axes]

    for i, name in enumerate(datasets):
        lora_scores = [results_data[name][n][0] for n in n_values if n in results_data[name]]
        reft_scores = [results_data[name][n][1] for n in n_values if n in results_data[name]]
        current_n = [n for n in n_values if n in results_data[name]]

        axes[i].plot(current_n, lora_scores, marker='o', label='LoRA', color='blue')
        axes[i].plot(current_n, reft_scores, marker='s', label='ReFT', color='green')
        axes[i].set_title(f"Accuracy Scaling: {name.upper()}")
        axes[i].set_xlabel("Sample Size (N)")
        axes[i].set_ylabel("Accuracy")
        axes[i].legend()
        axes[i].grid(True, linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.savefig(output_path)

def run_evaluation_harness():
    model_id = "meta-llama/Meta-Llama-3-8B"
    tokenizer = AutoTokenizer.from_pretrained(model_id)
    tokenizer.model_max_length = 2048
    tokenizer.pad_token = tokenizer.eos_token

    bnb_config = BitsAndBytesConfig(
        load_in_4bit=True,
        bnb_4bit_compute_dtype=COMPUTE_DTYPE,
        bnb_4bit_quant_type="nf4",
        bnb_4bit_use_double_quant=True
    )

    model = AutoModelForCausalLM.from_pretrained(model_id, quantization_config=bnb_config, device_map="auto", torch_dtype=COMPUTE_DTYPE)
    model = prepare_model_for_kbit_training(model, use_gradient_checkpointing=False)

    dp, datasets = DataProcessor(), DataProcessor.load_all()
    results_collector = {}
    pdf = FPDF(); pdf.add_page(); pdf.set_font("Arial", size=10)

    for name, ds in datasets.items():
        print(f"\n--- EVALUATING: {name.upper()} ---")
        results_collector[name] = {}
        for n in [16, 32, 64, 128, 256]:
            if n >= len(ds): continue

            context = [f"Input: {dp.get_fields(name, ds[i])[0]}\nOutput: {dp.get_fields(name, ds[i])[1]}" for i in range(n)]
            q_text, t_text = dp.get_fields(name, ds[n])
            query_str = f"Input: {q_text}\nOutput:"

            l_res = BaselineExecutor(model, tokenizer).run(query_str, context)
            r_res = ReFTExecutor(model, tokenizer).run(query_str, context)

            l_acc = dp.calculate_accuracy(l_res, t_text)
            r_acc = dp.calculate_accuracy(r_res, t_text)
            results_collector[name][n] = (l_acc, r_acc)

            log = f"[{name.upper()} N={n}] LoRA: {l_acc:.0%} | ReFT: {r_acc:.0%}"
            print(log)
            pdf.cell(0, 10, txt=log, ln=True)

    chart_path = "/content/drive/MyDrive/scaling_plot.png"
    generate_performance_chart(results_collector, chart_path)
    pdf.output("/content/drive/MyDrive/PyReft_stratified_report.pdf")

if __name__ == "__main__":
    run_evaluation_harness()
```
